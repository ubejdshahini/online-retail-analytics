## Setup

```python
import pandas as pd
import sys, os
sys.path.append(os.path.abspath('..'))


from src.data_cleaning import clean_data
from src.analysis import (
    get_monthly_revenue, get_revenue_by_hour, get_revenue_by_day_of_week,
    get_top_products, get_product_return_rate,
    get_country_performance, get_customer_lifetime_value,
    get_new_vs_returning_customers, get_kpi_summary,
)
from src.recommendation_engine import compute_rfm, get_segment_summary
from src.visualisations import (
    plot_monthly_revenue, plot_revenue_by_day_of_week, plot_revenue_by_hour,
    plot_top_products, plot_product_return_rates,
    plot_rfm_segments, plot_segment_revenue_share, plot_clv_distribution,
    plot_new_vs_returning, plot_country_revenue, plot_top_countries_bar,
    format_kpi_cards,
)
```

## Load & Clean Data

```python
import os
DATA_PATH = '../data/cleaned_retail_data.csv'
if not os.path.exists(DATA_PATH):
    print(f"File not found at {DATA_PATH}. Please run 01_data_exploration.ipynb first.")
else:
    print("Loading cleaned dataset...")
    df_clean = pd.read_csv(DATA_PATH, parse_dates=['InvoiceDate'])
    print(f"Cleaned shape: {df_clean.shape}")
    display(df_clean.head())
```

---
## 1. KPI Summary


```python
kpis = get_kpi_summary(df_clean)
cards = format_kpi_cards(kpis)
for c in cards:
    print(f"  {c['label']:<25} {c['value']}")
```

---
## 2. Revenue Trends

### 2a. Monthly Revenue + MoM Growth


```python
monthly = get_monthly_revenue(df_clean)
fig = plot_monthly_revenue(monthly)
fig.show()
```

### 2b. Revenue by Day of Week

```python
dow = get_revenue_by_day_of_week(df_clean)
fig = plot_revenue_by_day_of_week(dow)
fig.show()
```

### 2c. Revenue by Hour of Day

```python
hourly = get_revenue_by_hour(df_clean)
fig = plot_revenue_by_hour(hourly)
fig.show()
```

---
## 3. Product Performance

### 3a. Top 10 Products by Revenue


```python
top_products = get_top_products(df_clean, n=10)
fig = plot_top_products(top_products, n=10)
fig.show()
```

### 3b. Product Return Rates

```python
return_rates = get_product_return_rate(df_clean)
fig = plot_product_return_rates(return_rates, n=15)
fig.show()
```

---
## 4. Customer Analysis

### 4a. RFM Segment Bubble Chart


```python
rfm = compute_rfm(df_clean)
seg_summary = get_segment_summary(rfm)
fig = plot_rfm_segments(seg_summary)
fig.show()
```

### 4b. Revenue Share by Segment (Donut)

```python
fig = plot_segment_revenue_share(seg_summary)
fig.show()
```

### 4c. Customer Lifetime Value Distribution

```python
clv = get_customer_lifetime_value(df_clean)
fig = plot_clv_distribution(clv)
fig.show()
```

### 4d. New vs Returning Customers per Month

```python
new_ret = get_new_vs_returning_customers(df_clean)
fig = plot_new_vs_returning(new_ret)
fig.show()
```

---
## 5. Geographic Performance

### 5a. World Map (excl. UK)


```python
geo = get_country_performance(df_clean)
fig = plot_country_revenue(geo, exclude_uk=True)
fig.show()
```

### 5b. Top 10 Countries Bar Chart

```python
fig = plot_top_countries_bar(geo, n=10, exclude_uk=True)
fig.show()
```
